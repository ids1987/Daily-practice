Catagorical Features

  * Category data type -
  * Ordinal data type - you can set up a scale in the categories - Small       Medium Large - COld warm hot
  * small height medium height large height

Numerical Features

* where there are numbers
* will include float and integers both
* continous values,
* grouping up will not make much sense (unless doing clustering)

## Task exercise
* apply the three method to island column
* manually applying label codes
* using the factorize method
* using the category
* using the label encoder method

```python
#1️⃣ Manual label encoding (mapping dictionary)
island_map = {
    'Torgersen': 0,
    'Biscoe': 1,
    'Dream': 2
}

df['island_manual_encoded'] = df['island'].map(island_map)
```

```python
df[['island', 'island_manual_encoded']].head()
```

```python
#2️⃣ Encoding using factorize()
df['island_factorized'] = pd.factorize(df['island'])[0]
df
```

```python
#3️⃣ Encoding using category
df['island'] = df['island'].astype('category')
df['island_category_encoded'] = df['island'].cat.codes
```

```python
df[['island', 'island_category_encoded']].head()
```

```python
#Compare all 3 side-by-side
df[['island',
    'island_manual_encoded',
    'island_factorized',
    'island_category_encoded']].head(10)
```
