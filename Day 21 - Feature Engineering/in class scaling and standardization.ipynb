* Normalization

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

data = pd.DataFrame({
    'age': [20, 50, 100],
    'income': [20000, 50000, 100000]
})

# upscale/downscale
scaler = MinMaxScaler()

# Normalization
normalize_data = scaler.fit_transform(data)

df_normalized = pd.DataFrame(
    normalize_data,
    columns=['age', 'income']
)

df_normalized
```

Why do we normalize data?

easier to calculate
a lot prefernce to salary over age in the given example

our models only use numbers



```python
df_normalized.median()
```

```python
std_scaler = StandardScaler()

std_data = std_scaler.fit_transform(data)
std_data
```
