```python
iris_onehot = iris.copy()

encoder = OneHotEncoder( sparse_output = False, handle_unknown='ignore')

catagorical_cols = ['species']

ecoded_values = encoder.fit_transform(iris_onehot[catagorical_cols]) # species

# names of new columns, that don't exist yet
new_columns = encoder.get_feature_names_out(catagorical_cols)
new_columns
# ['species_setosa', 'species_versicolor', 'species_virginica']

# iris_onehot

# adding the columns to the dataframe

iris_onehot[new_columns] = ecoded_values

iris_onehot.drop(columns=['species'], inplace = True)
iris_onehot
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# copy dataframe
iris_onehot = iris.copy()
# create encoder
encoder = OneHotEncoder(
    sparse_output=False,
    handle_unknown='ignore'
)
# categorical column
categorical_cols = ['species']
# fit & transform
encoded_array = encoder.fit_transform(
    iris_onehot[categorical_cols]
)
# convert to DataFrame
encoded_df = pd.DataFrame(
    encoded_array,
    columns=encoder.get_feature_names_out(categorical_cols),
    index=iris_onehot.index
)
# drop original column & join encoded columns
iris_onehot = iris_onehot.drop(columns=categorical_cols)
iris_onehot = pd.concat([iris_onehot, encoded_df], axis=1)

iris_onehot
```
