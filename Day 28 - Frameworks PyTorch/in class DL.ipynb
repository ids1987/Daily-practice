```python
# guna data iris
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
# Load Iris
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1, 1)
# Standardize features
```

```python
# binarize to 2 classes (setosa and versicolor)
X = X[y.flatten() != 2]
y = y[y.flatten() != 2]
```

```python
# scaling
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        self.W1 = np.random.randn(input_size, hidden_size)
        self.b1 = np.zeros((1, hidden_size))
        self.W2 = np.random.randn(hidden_size, output_size)
        self.b2 = np.zeros((1, output_size))

    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))

    def sigmoid_derivative(self, x):
        return x * (1 - x)

    def feedforward(self, X):
        self.z1 = np.dot(X, self.W1) + self.b1
        self.a1 = self.sigmoid(self.z1)
        self.z2 = np.dot(self.a1, self.W2) + self.b2
        output = self.sigmoid(self.z2)
        return output

    def train(self, X, y, epochs, learning_rate):
        for _ in range(epochs):
            output = self.feedforward(X)

            error = y - output
            d_output = error * self.sigmoid_derivative(output)

            error_hidden = d_output.dot(self.W2.T)
            d_hidden = error_hidden * self.sigmoid_derivative(self.a1)

            self.W2 += self.a1.T.dot(d_output) * learning_rate
            self.b2 += np.sum(d_output, axis=0, keepdims=True) * learning_rate
            self.W1 += X.T.dot(d_hidden) * learning_rate
            self.b1 += np.sum(d_hidden, axis=0, keepdims=True) * learning_rate
```

```python
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([[0],
              [1],
              [1],
              [0]])

nn = NeuralNetwork(input_size=2, hidden_size=4, output_size=1)
nn.train(X, y, epochs=10000, learning_rate=0.1)

output = nn.feedforward(X)
print("Predictions after training:")
print(output)
```
