```python
import pandas as pd

df = pd.read_csv("malaysian_economic.csv")
```

```python
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])  #data cleaning
df = df.sort_values('date')    #sort time
df = df.dropna(subset=['date', 'leading'])  #missing value handling
```

```python
# create decade column
df['decade'] = (df['date'].dt.year // 10) * 10
# calculate average leading indicator by decade
decade_avg = df.groupby('decade')['leading'].mean().reset_index()
decade_avg
```

```python
#1st visual convert decade to string (categorical)
#To treat decades as categorical variables for clearer comparison
decade_avg['decade'] = decade_avg['decade'].astype(str)

hv.Bars(decade_avg, 'decade', 'leading').opts(
    title="Average Leading Economic Indicator by Decade (Malaysia)",
    width=700,
    height=400,
    xlabel="Decade",
    ylabel="Average Leading Index"
)
```

```python
# 2nd Visual -Leading vs coincident indicator
hv.Curve(df, 'date', 'leading').opts(
    title="Leading Economic Indicator Trend Over Time (Malaysia)",
    width=800,
    height=400,
    xlabel="Date",
    ylabel="Leading Index"
)







```
